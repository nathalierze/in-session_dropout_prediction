```python
import pandas as pd
import pickle
import seaborn as sns
import numpy as np
```

```python
infile = open('accuracyDTE.pkl','rb')
DTE = pickle.load(infile)
infile.close()

infile = open('accuracyKNN.pkl','rb')
KNN = pickle.load(infile)
infile.close()

infile = open('accuracylogReg.pkl','rb')
logreg = pickle.load(infile)
infile.close()

infile = open('accuracySVM.pkl','rb')
SVM = pickle.load(infile)
infile.close()

infile = open('accuracyDL.pkl','rb')
DL = pickle.load(infile)
infile.close()


DTE['model'] = 'DTE'
logreg['model'] = 'logreg'
KNN['model'] = 'KNN'
SVM['model'] = 'SVM'
DL['model'] = 'MLP'

df = DTE.append(KNN)
df = df.append(logreg)
df = df.append(SVM)
df = df.append(DL)
```

```python
df = df.reset_index()
```

```python
df.to_excel('metrics.xlsx')
```

```python
ax = sns.lineplot(data=df, x='Sentence', y='Accuracy', hue='model', palette="dark:#5A9_r")
```
