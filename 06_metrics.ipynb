```python
import pandas as pd
import pickle
import seaborn as sns
import warnings

# Ignore the warning message
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
# load metrics

infile = open("model_metrics/metrics_DTE.pkl", "rb")
DTE = pickle.load(infile)
infile.close()

infile = open("model_metrics/metrics_KNN.pkl", "rb")
KNN = pickle.load(infile)
infile.close()

infile = open("model_metrics/metrics_logistic_regr.pkl", "rb")
logreg = pickle.load(infile)
infile.close()

infile = open("model_metrics/metrics_DL.pkl", "rb")
DL = pickle.load(infile)
infile.close()


DTE["model"] = "DTE"
logreg["model"] = "logreg"
KNN["model"] = "KNN"
DL["model"] = "MLP"

df = DTE.append(KNN)
df = df.append(logreg)
df = df.append(DL)
df = df.reset_index()

# save combined results
# df.to_excel("metrics.xlsx")
```

```python
ax = sns.lineplot(data=df, x="Sentence", y="Accuracy", hue="model")
```
