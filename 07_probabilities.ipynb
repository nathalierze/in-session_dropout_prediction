```python
import pandas as pd
import pickle
import seaborn as sns
import numpy as np
import warnings
import matplotlib.pyplot as plt

# Ignore the warning message
warnings.filterwarnings("ignore", category=FutureWarning)
```

File loads probability data from sample n and plots it

```python
infile = open("probability_data/DTE5.pkl", "rb")
DTE5 = pickle.load(infile)
infile.close()
infile = open("probability_data/DTE15.pkl", "rb")
DTE15 = pickle.load(infile)
infile.close()
infile = open("probability_data/DTE25.pkl", "rb")
DTE25 = pickle.load(infile)
infile.close()
infile = open("probability_data/DTE35.pkl", "rb")
DTE35 = pickle.load(infile)
infile.close()
infile = open("probability_data/DTE45.pkl", "rb")
DTE45 = pickle.load(infile)
infile.close()
infile = open("probability_data/DTE55.pkl", "rb")
DTE55 = pickle.load(infile)
infile.close()

infile = open("probability_data/logreg5.pkl", "rb")
logreg5 = pickle.load(infile)
infile.close()
infile = open("probability_data/logreg15.pkl", "rb")
logreg15 = pickle.load(infile)
infile.close()
infile = open("probability_data/logreg25.pkl", "rb")
logreg25 = pickle.load(infile)
infile.close()
infile = open("probability_data/logreg35.pkl", "rb")
logreg35 = pickle.load(infile)
infile.close()
infile = open("probability_data/logreg45.pkl", "rb")
logreg45 = pickle.load(infile)
infile.close()
infile = open("probability_data/logreg55.pkl", "rb")
logreg55 = pickle.load(infile)
infile.close()


infile = open("probability_data/DL5.pkl", "rb")
dl5 = pickle.load(infile)
infile.close()
infile = open("probability_data/DL15.pkl", "rb")
dl15 = pickle.load(infile)
infile.close()
infile = open("probability_data/DL25.pkl", "rb")
dl25 = pickle.load(infile)
infile.close()
infile = open("probability_data/DL35.pkl", "rb")
dl35 = pickle.load(infile)
infile.close()
infile = open("probability_data/DL45.pkl", "rb")
dl45 = pickle.load(infile)
infile.close()
infile = open("probability_data/DL55.pkl", "rb")
dl55 = pickle.load(infile)
infile.close()
```

```python
DTE5["t"] = 5
DTE15["t"] = 15
DTE25["t"] = 25
DTE35["t"] = 35
DTE45["t"] = 45
DTE55["t"] = 55

dl5["t"] = 5
dl15["t"] = 15
dl25["t"] = 25
dl35["t"] = 35
dl45["t"] = 45
dl55["t"] = 55

logreg5["t"] = 5
logreg15["t"] = 15
logreg25["t"] = 25
logreg35["t"] = 35
logreg45["t"] = 45
logreg55["t"] = 55

ls = [
    DTE5,
    DTE15,
    DTE25,
    DTE35,
    DTE45,
    DTE55,
    logreg5,
    logreg15,
    logreg25,
    logreg35,
    logreg45,
    logreg55,
    dl5,
    dl15,
    dl25,
    dl35,
    dl45,
    dl55,
]

for x in ls:
    x.reset_index(inplace=True)
    x = x.rename(columns={"index": "index"})
```

```python
df = DTE5.append(DTE15)
df = df.append(DTE25)
df = df.append(DTE35)
df = df.append(DTE45)
df = df.append(DTE55)
df["model"] = "DTE"

df2 = dl5.append(dl15)
df2 = df2.append(dl25)
df2 = df2.append(dl35)
df2 = df2.append(dl45)
df2 = df2.append(dl55)
df2["model"] = "MLP"

df3 = logreg5.append(logreg15)
df3 = df3.append(logreg25)
df3 = df3.append(logreg35)
df3 = df3.append(logreg45)
df3 = df3.append(logreg55)
df3["model"] = "logreg"

df = df.append(df2)
df = df.append(df3)
```

```python
plt.figure(figsize=(10, 10))
sns.set(font_scale=1.5)
sns.set_style("white")

ax = sns.violinplot(x="t", y=0, data=df, hue="model")
ax.set(xlabel="Sentence", ylabel="P")
```
